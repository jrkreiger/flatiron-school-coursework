## Create the best plot using polynomials!

Here is a plot with a clearly non-linear shape. Begin to think about what degree polynomial you believe will fit it best.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

def pol(x):
    return x * np.cos(x)

x = np.linspace(0, 12, 100)
rng = np.random.RandomState(1234)
rng.shuffle(x)
x = np.sort(x[:25])
y = pol(x) + np.random.randn(25)*2

plt.scatter(x, y, color='green', s=50, marker='.')

plt.show();
```

Now it's time to determine the optimal degree of polynomial features for a model that is fit to this data     
- Plot a polynomial function using `PolynomialFeatures` for polynomials up until the second, third and fourth degrees. Make sure to label each degree polynomial in your plot.   
- Print out the $R^2$ value for each of the three results, then decide which degree is best.

## Solution

```python
# Transform the data to second, third, fourth degrees
poly2 = PolynomialFeatures(2)
x2 = poly2.fit_transform(x.reshape(-1,1))
display(x2[:5])

poly3 = PolynomialFeatures(3)
x3 = poly3.fit_transform(x.reshape(-1,1))
display(x3[:5])

poly4 = PolynomialFeatures(4)
x4 = poly4.fit_transform(x.reshape(-1,1))
display(x4[:5])
```

```python
# Fit linear regression models
reg2 = LinearRegression().fit(x2, y)
reg3 = LinearRegression().fit(x3, y)
reg4 = LinearRegression().fit(x4, y)
```

```python
# Calculate R2 for each model
print('x2 MSE:', mean_squared_error(y, reg2.predict(x2)))
print('x2 R2:', r2_score(y, reg2.predict(x2)))

print('x3 MSE:', mean_squared_error(y, reg3.predict(x3)))
print('x3 R2:', r2_score(y, reg3.predict(x3)))

print('x4 MSE:', mean_squared_error(y, reg4.predict(x4)))
print('x4 R2:', r2_score(y, reg4.predict(x4)))
```

```python
# Plot the data and the three models
plt.figure(figsize=(16,10))
plt.scatter(x, y, color='k')
plt.plot(x, reg2.predict(x2), color='magenta', label='2nd order')
plt.plot(x, reg3.predict(x3), color='aqua', label='3rd order')
plt.plot(x, reg4.predict(x4), color='darkorchid', label='4th order')
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0,15)
plt.title('Comparing three polynomial regression models')
plt.legend()
plt.show();
```

The fourth-degree polynomial model has the best fit.
