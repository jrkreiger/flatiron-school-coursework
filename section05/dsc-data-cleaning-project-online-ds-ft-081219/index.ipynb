Next, slice the DataFrame as needed and visualize the distribution of heights and weights by gender.  You should have 4 total plots.  

In the cell below:

* Slice the DataFrame into separate DataFrames by gender
* Complete the `show_distplot` function.  This helper function should take in a DataFrame, a string containing the gender we want to visualize, and the column name we want to visualize by gender. The function should display a distplot visualization from seaborn of the column/gender combination.  

Hint: Don't forget to check the [seaborn documentation for distplot](https://seaborn.pydata.org/generated/seaborn.distplot.html) if you have questions about how to use it correctly! 

```python
def show_distplot(dataframe, gender, column_name):
    data = dataframe[dataframe['Gender'] == gender][column_name]
    sns.distplot(data)
    plt.title('Distribution of {} for {} Superheroes'.format(column_name, gender))
    plt.ylabel('Probability Density Function')
    print('Mean {} for {} Superheroes: {}'.format(column_name, gender, data.mean()))
    print('Median {} for {} Superheroes: {}'.format(column_name, gender, data.median()))
```

```python
# Male Height
show_distplot(heroes_df, 'Male', 'Height')
```

```python
# Male Weight
show_distplot(heroes_df, 'Male', 'Weight')
```

```python
# Female Height
show_distplot(heroes_df, 'Female', 'Height')
```

```python
# Female Weight
show_distplot(heroes_df, 'Female', 'Weight')
```

```python
def top_5_powers(dataframe):
    df = dataframe.drop(heroes_df.columns.values[1:], axis=1)
    columns = df.columns.values
    for col in columns:
        df[col] = df[col].map({"True": 1, "False": 0})
        
    power_counts_dict = dict(df.sum())
    
    return sorted(power_counts_dict.items(), key=lambda x: x[1], reverse=True)[:5]
 
top_5_powers(heroes_and_powers)
```

```python
marvel_df = heroes_and_powers[heroes_and_powers['Publisher'] == 'Marvel Comics']
top_5_powers(marvel_df)
```

```python
dc_df = heroes_and_powers[heroes_and_powers['Publisher'] == 'DC Comics']
top_5_powers(dc_df)
```

### Just your average superhero

The [Super Heroes Dataset](https://www.kaggle.com/claudiodavi/superhero-set/data) represents characters in the Marvel and DC Comic Universes. According to the data from these comics, what does the average superhero look like?

Above we learned that 71% of superheros are male characters with a median height of 183 cm and median weight of 79 kg. The dataset also contains information on hair, skin, and eye colors. Let's explore it to complete our picture of the average superhero.


```python
# Create a function to replace hyphens with 'unspecified' for missing values
def relabel_unspecified(dataframe, col_name):
    dataframe[col_name].replace('-', 'unspecified', inplace = True)

# Print counts of all the eye colors in the dataset
relabel_unspecified(heroes_df, 'Eye color')
heroes_df['Eye color'].value_counts(normalize=True)
```

```python
# Print counts of all the hair colors in the dataset
relabel_unspecified(heroes_df, 'Hair color')
heroes_df['Hair color'].value_counts(normalize=True)
```

```python
# Print counts of all the skin colors in the dataset
relabel_unspecified(heroes_df, 'Skin color')
heroes_df['Skin color'].value_counts(normalize=True)
```

```python
# Plot top 10 eye colors
heroes_df['Eye color'].value_counts(normalize=True)[:10].plot(kind='barh', color='black')
plt.xlabel('Percent')
plt.title('Top 10 Hero Eye Colors');
```

```python
# Plot top 10 hair colors
heroes_df['Hair color'].value_counts(normalize=True)[:10].plot(kind='barh', color='black')
plt.xlabel('Percent')
plt.title('Top 10 Hero Hair Colors');
```

```python
# Plot top 10 skin colors
heroes_df['Skin color'].value_counts(normalize=True)[:10].plot(kind='barh', color='black')
plt.xlabel('Percent')
plt.title('Top 10 Hero Skin Colors');
```

As the tables and charts above show, skin color is unspecified in 90% of cases. Nine percent of heroes have what we might call an "alien" skin color (green, blue, red, purple, etc.), while only 1% of heroes are described as having white or black skin, and none brown. (It's also not clear from the data if "white" and "black" refer to human skin tones at all.) Hair color is unspecified for 23% of heroes, but almost just as often it is black (22%). Blue eyes are the most common (31%), followed by unspecified (23%), then brown (17%), green (10%), and on down a list of unusual colors.

Without any exposure to the source material, we could surmise from the data that the average superhero in the DC/Marvel Universes is a male character with blue eyes. He is more likely to have black hair than any other color, and his skin color is left up to our imaginations. [Sound familiar?](https://en.wikipedia.org/wiki/Superman)
