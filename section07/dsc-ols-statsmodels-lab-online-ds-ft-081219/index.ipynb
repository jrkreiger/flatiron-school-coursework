## Step 1: Read the dataset and inspect its columns and 5-point statistics

```python
# Load necessary libraries and import the data
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
```

```python
# Import data
df = pd.read_csv('advertising.csv', index_col=0)
```

```python
# Check the columns and first few rows
df.head()
```

```python
# Get the 5-point statistics for data 
df.describe()
```

The dataset contains four columns: 
- three represent advertising channels: TV, radio, newspaper
- the fourth represents sales.

Units are not stated in the documentation.

## Step 2: Plot histograms with kde overlay to check the distribution of the predictors

```python
# For all the variables, check distribution by creating a histogram with kde
sns.distplot(df.TV);
```

```python
sns.distplot(df.radio);
```

```python
sns.distplot(df.newspaper);
```

```python
sns.distplot(df.sales);
```

TV and radio are far from normally distributed (they each have two peaks in their KDE functions). Newspaper is right-skewed. Sales is close to normally distributed.


## Step 3: Test for the linearity assumption 

Use scatterplots to plot each predictor against the target variable

```python
# visualize the relationship between the preditors and the target using scatterplots
fig = plt.figure(figsize=(16,6))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

ax1.scatter(df.TV, df.sales, label='TV')
ax1.set_title('TV v. sales')
ax1.set_xlabel('TV')
ax1.set_ylabel('Sales')

ax2.scatter(df.radio, df.sales, label='radio')
ax2.set_title('Radio v. sales')
ax2.set_xlabel('Radio')
ax2.set_ylabel('Sales')

ax3.scatter(df.newspaper, df.sales, label='newspaper')
ax3.set_title('Newspaper v. sales')
ax3.set_xlabel('Newspaper')
ax3.set_ylabel('Sales')

plt.show();
```

TV and radio appear to have linear relationships with sales, but newspaper does not.

### Conclusion so far

Based on above initial checks, we can confidently say that TV and radio appear to be good predictors for our regression analysis. Newspaper is very heavily skewed and also doesnt show any clear linear relationship with the target. 
> We'll move ahead with our analysis using TV and radio, and rule out newspaper because we believe it violates OLS assumptions

Note: Kurtosis can be dealt with using techniques like log normalization to "push" the peak towards the center of distribution. You'll learn about this later on.

## Step 4: Run a simple regression in Statsmodels with TV as a predictor

```python
# build the formula 
f = 'sales~TV'

# create a fitted model in one line
model = ols(formula=f, data=df).fit()
```

## Step 5: Get Regression Diagnostics Summary

```python
model.summary()
```

Note here that the coefficients represent associations, not causations

## Step 6:  Draw a prediction line with data points on a scatter plot for X (TV) and Y (Sales)

Hint: You can use the `model.predict()` function to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV
new_x = pd.DataFrame({'TV': [df.TV.min(), df.TV.max()]})

# make predictions for those x values and store them
predicted = model.predict(new_x)

# first, plot the observed data and the least squares line
plt.scatter(df.TV, df.sales, label='Data')
plt.plot(new_x, predicted, color='red', label='Prediction line')

plt.legend()
plt.show();
```

## Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig = plt.figure(figsize=(16,8))
fig = sm.graphics.plot_regress_exog(model, 'TV', fig=fig);
```

There is a generally conical shape to the distribution of the residuals, which means our data may not be homoscedastic. This is turn means that linear regression may not have been an appropriate sort of model for this data.

## Step 8: Check the normality assumptions by creating a QQ-plot

```python
# Code for QQ-plot here
import scipy.stats as stats
fig = sm.graphics.qqplot(model.resid, dist=stats.norm, line='45', fit=True)
plt.show();
```

Based on the Q-Q plot, the residuals appear to be normally distributed.

## Step 9: Repeat the above for radio and record your observations

```python
# code for model, prediction line plot, heteroscedasticity check and QQ normality check here
```

```python
# build the formula 
f2 = 'sales~radio'

# create a fitted model in one line
model2 = ols(formula=f2, data=df).fit()
```

```python
model2.summary()
```

```python
# create a DataFrame with the minimum and maximum values of radio
new_x2 = pd.DataFrame({'radio': [df.radio.min(), df.radio.max()]})

# make predictions for those x values and store them
predicted2 = model2.predict(new_x2)

# first, plot the observed data and the least squares line
plt.scatter(df.radio, df.sales, label='Data')
plt.plot(new_x2, predicted2, color='red', label='Prediction line')

plt.legend()
plt.show();
```

```python
# Check for goodness of fit and heterscedasticity
fig = plt.figure(figsize=(16,8))
fig = sm.graphics.plot_regress_exog(model2, 'radio', fig=fig);
```

```python
# Code for QQ-plot here
fig = sm.graphics.qqplot(model2.resid, dist=stats.norm, line='45', fit=True)
plt.show();
```

Uh-oh! Looks like the residuals are heteroscedastic and not normally distributed.
