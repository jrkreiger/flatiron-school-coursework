```python
# Import needed packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc, confusion_matrix

import statsmodels.api as sm
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

## Problem Formulation

Define X and y as with the previous lab. This time, follow best practices and also implement a standard train-test split.  
**For consistency of results, use random_state=17.**

```python
# Define X and y
X = df.drop('target', axis=1)
y = df['target']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=17)
```

## Initial Model - Personal Implementation

Use your code from the previous lab to once again train a logistic regression algorithm on the training set.

```python
# Define logistic regression function
def sig_fun(x):
    denom = 1 + np.exp(-x)
    return 1/denom

def grad_desc(X, y, max_iterations, alpha, initial_weights=None):
    """Be sure to set default behavior for the initial_weights parameter."""
    if initial_weights == None:
        initial_weights = np.ones((X.shape[1],1)).flatten()
    weights_col = pd.DataFrame(initial_weights)
    weights = initial_weights
    
    #Create a for loop of iterations
    for it in range(max_iterations):
        
        #Generate predictions using the current feature weights
        y_pred = sig_fun(X.dot(weights))
        
        #Calculate an error vector based on these initial predictions and the correct labels
        errors = y - y_pred
        
        #Calculate the gradient       
        #As we saw in the previous lab, calculating the gradient is often the most difficult task.
        #Here, your are provided with the closed form solution for the gradient of the log-loss function derived from MLE
        #For more details on the derivation, see the additional resources section below.
        gradient = np.dot(X.transpose(),errors) 
        
        #Update the weight vector take a step of alpha in direction of gradient 
        weights += alpha*gradient
        weights_col = pd.concat([weights_col, pd.DataFrame(weights)], axis=1)
        
    #Return finalized Weights
    return weights, weights_col
```

```python
weights, weights_col = grad_desc(X_train, y_train, max_iterations=10000, alpha=0.001)
```

## Now use your algorithm to make [probability] predictions on the test set

```python
# Calculate prob of target=1 using sigmoid function and model coefficients
y_pred = sig_fun(X_test.dot(weights))
y_pred = [round(pred,3) for pred in y_pred]
y_pred[:5]
```

## Create an ROC curve for your predictions

```python
# Calculate fpr, tpr
fpr, tpr, thresholds = roc_curve(y_test, y_pred)

# Plot ROC curve
plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, label='ROC curve')
plt.plot([0,1],[0,1], linestyle='--')
plt.title('ROC curve for my model')
plt.legend();

# Print AUC
print('AUC:', auc(fpr, tpr))
```

## Update your ROC curve to not only include a graph of the test set, but one of the train set

```python
# Calculate fpr, tpr for train set
y_pred_train = sig_fun(X_train.dot(weights))
y_pred_train = [round(pred,3) for pred in y_pred_train]

fpr_train, tpr_train, thresholds = roc_curve(y_train, y_pred_train)

# Plot ROC curve
plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, label='Test set')
plt.plot(fpr_train, tpr_train, label='Train set')
plt.plot([0,1],[0,1], linestyle='--')
plt.title('ROC curve for my model')
plt.legend();

# Print AUC
print('Test AUC:', auc(fpr, tpr))
print('Train AUC:', auc(fpr_train, tpr_train))
```

## Adding an Intercept

Now add an intercept to the scikit-learn model. Keep the regularization parameter C set to a very large number such as 1e16. Plot all three models ROC curves on the same graph.

```python
# Add an intercept to the model
Xc = sm.add_constant(X)

Xc_train, Xc_test, y_train, y_test = train_test_split(Xc, y, random_state=17)

logreg2 = LogisticRegression(fit_intercept=True, C=1e16, solver='liblinear')
model2 = logreg2.fit(Xc_train, y_train)

# Calculate predictions, fpr, tpr
y_score_Xc = model2.decision_function(Xc_test)

fpr_Xc, tpr_Xc, thresholds = roc_curve(y_test, y_score_Xc)

# Plot ROC curve
plt.figure(figsize=(10,8))
plt.plot(fpr, tpr, label='Test set')
plt.plot(fpr_train, tpr_train, label='Train set')
plt.plot(fpr_Xc, tpr_Xc, label='Test with intercept')
plt.plot([0,1],[0,1], linestyle='--', color='k')
plt.title('ROC curve for sklearn model')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xticks([round(i,2) for i in np.arange(0,1.05,0.05)])
plt.yticks([round(i,2) for i in np.arange(0,1.05,0.05)])
plt.legend(loc=4);


# Print AUC
print('Test AUC:', auc(fpr, tpr))
print('Train AUC:', auc(fpr_train, tpr_train))
print('AUC for model with intercept:', auc(fpr_Xc, tpr_Xc))
```

## Altering the Regularization Parameter

Now, experiment with altering the regularization parameter. At a minimum, create 5 different subplots with varying regularization (C) parameters. For each, plot the ROC curve of the train and test set for that specific model.  

Regularization parameters between 1 and 20 are recommended. Observe the difference in test and train auc as you go along.

```python
# Create list of regularization parameters
c_values = list(range(1,21))

# Initialize figure
plt.figure(figsize=(16,16))

# Plot ROC curves for each c value
for index, c in enumerate(c_values):
    
    # Create and fit model
    logreg = LogisticRegression(fit_intercept=False, C=1.5**c, solver='liblinear')
    model = logreg.fit(X_train, y_train)
    
    # Calculate params
    y_score = model.decision_function(X_test)
    y_score_train = model.decision_function(X_train)
    fpr, tpr, threshold = roc_curve(y_test, y_score)
    fpr_train, tpr_train, threshold = roc_curve(y_train, y_score_train)
    auc_test = round(auc(fpr,tpr),3)
    auc_train = round(auc(fpr_train, tpr_train),3)
    auc_diff = round((auc_test-auc_train),3)
    
    # Create subplot
    plt.subplot(5,4,index+1)
    plt.plot(fpr, tpr, label='Test set')
    plt.plot(fpr_train, tpr_train, label='Train set')
    plt.plot([0,1],[0,1], linestyle='--', color='k')
    plt.title('ROC curve for C=1.5^{}\nTest AUC - Train AUC = {}'.format(c, auc_diff))
    plt.text(0.675,0.3,'Test AUC: {}'.format(auc_test))
    plt.text(0.675,0.25,'Train AUC: {}'.format(auc_train))
    plt.xlim([0.0,1.0])
    plt.ylim([0.0,1.05])
    plt.xticks([round(i,2) for i in np.arange(0,1.05,0.2)])
    plt.yticks([round(i,2) for i in np.arange(0,1.05,0.2)])
    plt.legend(loc=4)
plt.tight_layout()
plt.show();
```

## Comment on how the Regularization Parameter Impacts the ROC curves plotted above

The difference in AUC between the training and test sites is at its lowest when C is lowest, and the difference increases to converge at -0.03 as C increases. As C increases, test AUC decreases and train AUC increases. This tells me that raising C (i.e., increasing regularization) leads to overfitting on the training set, resulting in reduced performance on the test set.
