The four main SQLite3 datatypes are:
- text
- integer
- real (i.e., decimal/float/double, up to 15 characters)
- blob (used for binary data)

We can use other common types from other dialects of SQL; if possible, they will be coerced to SQLite3 datatypes (e.g., INT will be coerced to INTEGER)



## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

Type your answer below this line:
_______________________________________________________________________________________________________________________________

The primary key is a column containing a unique value for each row in a table. There can be more than one primary key used in combination to uniquely identify each row.

A foreign key is the primary key of another table, which will not necessarily be a unique value in the table one is currently referencing.



## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________

One-to-one: a unique value in Table 1 matches a unique value in Table 2.
One-to-many: a unique value in Table 1 matches several values in Table 2.
Many-to-many: a recurring value in Table 1 matches several values in Table 2.


## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________

Inner join: selects rows where two tables have matching values in a certain column
Left join: selects all rows from Table 1 and any rows from Table 2 with matching values in a certain column
Right join: selects all rows from Table 2 and any rows from Table 1 with matching values in a certain column
Outer join: selects rows where two tables _do not_ have matching values in a certain column (not supported in SQLite3)


## Question 5

Explain the relationship between Aggregate functions and GROUP BY statements.

Type your answer below this line:
_______________________________________________________________________________________________________________________________

Aggregate functions are used in conjunction with GROUP BY statements to perform some calculation on data that has been grouped according to values in a certain column. Once data has been grouped by values in a certain column, an aggregate function can then be applied to calculate the mean, median, sum, count, etc., of values per group.


## Question 6

What role do Associative Entities play (JOIN Tables) in many-to-many JOINs?

In a many-to-many join, associative entities contain all possible combinations of selected values from Table 1 and Table 2. Each row in the JOIN table represents a unique combination of a value from Table 1 and a value from Table 2. The JOIN table is a third location outside of Tables 1 & 2.
