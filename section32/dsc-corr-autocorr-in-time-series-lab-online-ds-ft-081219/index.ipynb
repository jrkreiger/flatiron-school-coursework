## The Exchange Rate Data

We'll be looking at the exchange rates dataset again. First, import the necessary libraries for time series and plotting. Then import the data (in `exch_rates.csv`) and make sure it's set in the correct time series format with the `datetime` as the index.

```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
from matplotlib.pylab import rcParams
```

```python
# Import data
data = pd.read_csv('exch_rates.csv')
```

```python
data.head()
```

```python
data.Frequency = pd.to_datetime(data.Frequency, format='%Y-%m-%d')
data.set_index('Frequency', inplace=True)
data.info()
```

Plot the three exchange rates in one plot

```python
# Plot the three exchange rates
plt.figure(figsize=(16,6))
plt.plot(data.index, data['Euro'], label='Euro')
plt.plot(data.index, data['Australian Dollar'], label='Australian Dollar')
plt.plot(data.index, data['Danish Krone'], label='Danish Krone')
plt.legend();
```

You can see that the EUR/USD and AUD/USD exchange rate are somewhere between rougly 0.5 and 2 between 2000 and 2018, where the Danish Krone is somewhere between roughly 4.5 and 9. Now let's look at the correlations between these time series.

```python
# Check correlations of raw series
data.corr()
```

### What is your conclusion here? You might want to use outside resources to understand what's going on.

Euro and Danish Krone are almost perfectly correlated, which makes sense because [the value of Krone is tied to the same exchange rate mechanism as the Euro](https://en.wikipedia.org/wiki/Danish_krone). 

Next, look at the plots of the differenced series. Use subplots to plot them rather than creating just one plot.

```python
# Create differenced series
data_diff = data.diff()
```

```python
# Plot the difference series
data_diff.plot(figsize=(16,10), subplots=True, legend=True);
```

### Explain what's going on

Next, let's look at the "lag 1 autocorrelation" for the EUR/USD exchange rate. Create a "lag 1 autocorrelation" series, plot the result, and look at the correlation coefficient.

```python
# your code here
euro = data[['Euro']]
euro_shift = euro.shift(periods=1)

lag1 = pd.concat([euro_shift, euro], axis=1)
lag1.columns = ['Euro_lag', 'Euro']
display(lag1.corr())

lag1.plot(figsize=(16,6), subplots=False, legend=True, alpha=0.5);
```

The original Euro values and the Euro values shifted by 1 day are almost perfectly correlated (0.999).

Repeat this, but for a "lag 50 autocorrelation"

```python
euro_shift50 = euro.shift(periods=50)

lag50 = pd.concat([euro_shift50, euro], axis=1)
lag50.columns = ['Euro_lag50', 'Euro']
display(lag50.corr())

lag50.plot(figsize=(16,6), subplots=False, legend=True, alpha=0.5);
```

The correlation is a little less with a 50-day lag (0.968).

### What's your conclusion here?

Knowing this, let's plot the ACF now.

```python
# your code here
plt.figure(figsize=(16,6))
pd.plotting.autocorrelation_plot(euro.dropna());
```

The series is heavily autocorrelated at first, and then there is a decay. This is a typical result for a series that is a random walk, generally you'll see heavy autocorrelations first, slowly tailing off until there is no autocorrelation anymore.

Next, let's look at the Partial Autocorrelation Function.

```python
# your code here
rcParams['figure.figsize'] = 16,6
plot_pacf(euro.dropna(), lags=100, method='ywm');
```

This is interesting! Remember that *Partial Autocorrelation Function* gives the partial correlation of a time series with its own lagged values, controlling for the values of the time series at all shorter lags. When controlling for 1 period lags, the PACF is only very high for one-period lags, and basically 0 for shorter lags. This is again a typical result for Random Walk series!

## The Airpassenger Data

Look at ACF and PACF for the airpassenger data and describe the result `passengers.csv`. Do this both for the differenced and regular series.

### Regular series

```python
# your code here
pax = pd.read_csv('passengers.csv')
pax.Month = pd.to_datetime(pax.Month, format='%Y-%m-%d')
pax.set_index('Month', inplace=True)
pax.info()
```

```python
pax.plot(figsize=(16,6));
```

```python
# your code here
rcParams['figure.figsize'] = 16,6
plot_acf(pax.dropna(), lags=100);
```

```python
# your code here
plot_pacf(pax.dropna(), lags=100, method='ywm');
```

### Differenced series

```python
# your code here
pax_diff = pax.diff(periods=1)

pax_diff.plot(figsize=(16,6));
```

```python
# your code here
pd.plotting.autocorrelation_plot(pax_diff.dropna());

rcParams['figure.figsize'] = 16,6
plot_acf(pax_diff.dropna(), lags=100);
plot_pacf(pax_diff.dropna(), lags=100, method='ywm');
```

### Your conclusion here

Seasonality is more apparent in the differenced series; there are strong auto-correlations at intervals of 12 lags.
