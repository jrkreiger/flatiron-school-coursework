## Detrending the Airpassenger data

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Import passengers.csv and set it as a time-series object. Plot the TS
data = pd.read_csv('passengers.csv')
data.info()
```

```python
data.Month = pd.to_datetime(data.Month, format='%Y-%m')
data.set_index('Month', inplace=True)
data.head()
```

```python
data.plot(figsize=(16,6));
```

## Create a stationarity check

At this stage, we can use the code from previous labs to create a function `stationarity_check(ts)` that takes in a time series object and performs stationarity checks including rolling statistics and the Dickey Fuller test. 

We want the output of the function to:
- Plot the original time series along with the rolling mean and rolling standard deviation in one plot
- Output the results of the Dickey-Fuller test

```python
# Create a function to check for the stationarity of a given timeseries using rolling stats and DF test
# Collect and package the code from previous lab
def stationarity_check(data, col=None):
    rollmean = data.rolling(window=8, center=False).mean()
    rollstd = data.rolling(window=8, center=False).std()

    plt.figure(figsize=(16,6))
    plt.plot(data.index, data, label='Passengers')
    plt.plot(data.index, rollmean, label='Rolling mean')
    plt.plot(data.index, rollstd, label='Rolling st. dev.')
    plt.legend();

    from statsmodels.tsa.stattools import adfuller

    #Perform Dickey-Fuller test:
    if col == None:
        test = adfuller(data)
    else:
        test = adfuller(data[col])
    output = pd.Series(test[0:4], index=['Test stat.', 'p-value', '# Lags used',
                                        '# obs used'])

    # Extract and display test results in a user friendly manner
    for key, value in test[4].items():
        output['Crit. value (%s)'%key] = value

    print(output)
```

Use your newly created function on the airpassenger data set.

```python
# Code here
stationarity_check(data, '#Passengers')
```

## Perform a log() and sqrt() transform

```python
# Log transform timeseries and compare with original to check the effect
data_log = np.log(data)
data_sqrt = np.sqrt(data)

data.plot(figsize=(16,6), title='Original')
data_log.plot(figsize=(16,6), title='Log-transformed')
data_sqrt.plot(figsize=(16,6), title='Square-root transformed');
```

moving forward, let's keep working with the log transformed data before subtracting rolling mean, differencing, etc.

## Subtracting the rolling mean

Create a rolling mean using your log transformed time series, with a time window of 7. Plot the log-transformed time series and the rolling mean together.

```python
# your code here
rollmean = data_log.rolling(window=7, center=False).mean()

plt.figure(figsize=(16,6))
plt.plot(data_log.index, data_log, label='Passengers (log-transformed)')
plt.plot(data_log.index, rollmean, label='Rolling mean')
plt.legend();
```

Now, subtract the rolling mean from the time series, look at the 10 first elements of the result and plot the result.

```python
# Subtract the moving average from the original data and check head for Nans
data_sub_mean = data_log - rollmean
data_sub_mean.head(10)
```

```python
# Drop the NaN values from timeseries calculated above
data_sub_mean.dropna(inplace=True)
data_sub_mean.head(10)
```

```python
# Plot the result
plt.figure(figsize=(16,6))
plt.plot(data_sub_mean.index, data_sub_mean, label='Passengers (log-transformed and minus rolling mean)')
plt.legend();
```

Finally, use your function `check_stationarity` to see if this series is considered stationary!

```python
# Your code here
stationarity_check(data_sub_mean, '#Passengers')
```

### Based on the visuals and on the Dickey-Fuller test, what do you conclude?


It's still not stationary. The rolling standard deviation is increasing over time.

## Subtracting the weighted rolling mean

Repeat all the above for the *weighted* rolling mean. Start from the log-transformed data again. Compare the Dickey-Fuller Test results. What do you conclude?

```python
# Use Pandas ewma() to calculate Weighted Moving Average of ts_log
ewma = data_log.ewm(halflife=2).mean()

# Plot the original data with exp weighted average
plt.figure(figsize=(16,6))
plt.plot(data_log.index, data_log, label='Passengers (log-transformed and differenced)')
plt.plot(data_log.index, ewma, label='Exp. WMA')
plt.legend();
```

```python
# Subtract the moving average from the original data and plot
data_sub_ewma = data_log - ewma
```

```python
data_sub_ewma.plot(figsize=(16,6));
```

```python
# do a stationarity check
stationarity_check(data_sub_ewma, '#Passengers')
```

### Based on the visuals and on the Dickey-Fuller test, what do you conclude?


It looks like the best combination of preprocessing steps to get stationary data was to log-transform the original data and then subtract the exponentially weighted moving average.

## Differencing

Using exponentially weighted moving averages, we seem to have removed the upward trend, but not the seasonality issue. Now use differencing to remove seasonality. Make sure you use the right amount of `periods`. Start from the log-transformed, exponentially weighted rolling mean-subtracted series.

After you differenced the series, run the `stationarity check` again.

```python
# difference your data and look at the head
data_diff = data_sub_ewma.diff(periods=12)
data_diff.dropna(inplace=True)
data_diff.head(15)
```

```python
# plot your differenced time series
plt.figure(figsize=(16,6))
plt.plot(data_sub_ewma.index, data_sub_ewma, label='Passengers (log-trans. & EWMA)')
plt.plot(data_diff.index, data_diff, label='Differenced')
plt.legend();
```

```python
# perform the stationarity check
stationarity_check(data_diff,'#Passengers')
```

### Your conclusion

The best combination of preprocessing steps to remove time-based trends was to log-transform the original data, subtract the exponentially weighted moving average, and then difference the data over a 12-month period. The resulting data is stationary (p-value=0.0008 in the Dickey-Fuller test).
