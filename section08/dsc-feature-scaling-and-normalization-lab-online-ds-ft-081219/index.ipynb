## Back to our Boston Housing data

Let's import our Boston Housing data. Remember we categorized two variables and deleted the "NOX" (nitride oxide concentration) variable because it was highly correlated with two other features.

```python
import pandas as pd
import numpy as np
%matplotlib inline
from sklearn.datasets import load_boston
boston = load_boston()

boston_features = pd.DataFrame(boston.data, columns = boston.feature_names)

# first, create bins for RAD based on the values observed. 5 values will result in 4 bins
bins = [0, 3, 4 , 5, 24]
bins_rad = pd.cut(boston_features['RAD'], bins)
bins_rad = bins_rad.cat.as_unordered()

# first, create bins for TAX based on the values observed. 6 values will result in 5 bins
bins = [0, 250, 300, 360, 460, 712]
bins_tax = pd.cut(boston_features['TAX'], bins)
bins_tax = bins_tax.cat.as_unordered()

tax_dummy = pd.get_dummies(bins_tax, prefix="TAX", drop_first=True)
rad_dummy = pd.get_dummies(bins_rad, prefix="RAD", drop_first=True)
boston_features = boston_features.drop(["RAD","TAX"], axis=1)
boston_features = pd.concat([boston_features, rad_dummy, tax_dummy], axis=1)
boston_features = boston_features.drop("NOX",axis=1)
```

## Look at the histograms for the continuous variables

```python
# Inspect histograms
cols = ['CRIM', 'ZN', 'INDUS', 'RM', 'AGE', 'DIS', 'PTRATIO', 'B',
       'LSTAT']
boston_features[cols].hist(figsize=(16,12));
```

## Perform log transformations for the variables where it makes sense

Analyze the results in terms of how they improved the normality performance. What is the problem with the "ZN" variable?  

```python
# Log-transform 'ZN', 'AGE', and 'B'
zn_log = boston_features['ZN'].map(lambda x: np.log(x))
age_log = boston_features['AGE'].map(lambda x: np.log(x))
b_log = boston_features['B'].map(lambda x: np.log(x))

log_vars = pd.concat({'zn_log': zn_log, 'age_log': age_log, 'b_log': b_log}, axis=1)
log_vars.head()

cols = ['age_log', 'b_log']
log_vars[cols].hist(figsize=(16,6));
```

"ZN" has a lot of zeros (more than 50%!). Remember that this variable denoted: "proportion of residential land zoned for lots over 25,000 sq.ft.". It might have made sense to categorize this variable to "over 25,000 feet or not (binary variable 1/0). Now you have a zero-inflated variable which is cumbersome to work with.

```python
# Log-transform all the other columns except 'RM' and 'ZN'
data_log = pd.DataFrame([])

data_log['crim_log'] = np.log(boston_features['CRIM'])
data_log['indus_log'] = np.log(boston_features['INDUS'])
data_log['dis_log'] = np.log(boston_features['DIS'])
data_log['ptratio_log'] = np.log(boston_features['PTRATIO'])
data_log['lstat_log'] = np.log(boston_features['LSTAT'])
data_log['age_log'] = age_log
data_log['b_log'] = b_log


data_log.head()
```

```python
log_cols = ['crim_log', 'indus_log', 'dis_log', 'ptratio_log', 'lstat_log', 
            'age_log', 'b_log']
data_log[log_cols].hist(figsize=(16,12));
```

Log transformation seems to have improved skewness on most variables except AGE and B. When we do the next round of transformations, we'll use raw AGE and B from the boston_features DataFrame, and the log-transformed versions of the rest of the features.

## Try different types of transformations on the continuous variables

Store your final features in a dataframe `features_final`

```python
# Initialize each column for transformation
age = boston_features['AGE']
b = boston_features['B']
crim_log = data_log['crim_log']
dis_log = data_log['dis_log']
indus_log = data_log['indus_log']
lstat_log = data_log['lstat_log']
ptratio_log = data_log['ptratio_log']
rm = boston_features['RM']

# Initialize clean_features DataFrame
clean_features = pd.DataFrame([])

```

```python
# Transform each feature to increase normality

# Age: mean-normalization
clean_features['age'] = (age - np.mean(age)) / (max(age) - min(age))

# B: min-max scaling
clean_features['b'] = (b - min(b)) / (max(b) - min(b))

# Crim_log: min-max scaling
clean_features['crim'] = (crim_log - min(crim_log)) / (max(crim_log) - min(crim_log))

# Dis_log: standardization a.k.a. z-score normalization
clean_features['dis'] = (dis_log - np.mean(dis_log)) / np.sqrt(np.var(dis_log))

# Indus_log: standardization a.k.a. z-score normalization
clean_features['indus'] = (indus_log - np.mean(indus_log)) / np.sqrt(np.var(indus_log))

# Lstat_log: mean normalization
clean_features['lstat'] = (lstat_log - np.mean(lstat_log)) / (max(lstat_log) - min(lstat_log))

# Ptratio_log: unit vector transformation
clean_features['ptratio'] = ptratio_log / np.linalg.norm(ptratio_log)

# Rm: min-max scaling
clean_features['rm'] = (rm - min(rm)) / (max(rm) - min(rm))

clean_features.head()
```
