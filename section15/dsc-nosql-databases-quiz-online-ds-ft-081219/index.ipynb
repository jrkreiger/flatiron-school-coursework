## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

SQL databases are relational databases: they store data in tables that are related to one another through matching values.

NoSQL databases store data in other ways: as documents, graphs, key-value dictionaries, columns, or distributed across multiple server clusters.

Which is better depends on a business's needs. Relatively small amounts of data whose structure is well known or determined in advance will work well in a relational database. Big data, network data, and data whose structure is not known in advance or not easily translated to tables would do better in a NoSQL database.


## Question 2

Describe a situation where a NoSQL database might be a better choice than a relational database, and explain your reasoning.

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

A NoSQL database would be better than a relational one for data representing a network, such as a social network. To store social network data in tables, one would need to make long lists of every edge connecting nodes, and querying would be cumbersome. A graph database is much better suited to storing and querying data whose main purpose is to record sets of relationships.

## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

Fault tolerance is the ability of a database to remain useful for querying even if part of the storage system fails. When a dataset is spread across multiple servers, a failure in one server means that some of the data may no longer be available for querying (and we would say that such a dataset is not fault tolerant). Resilient distributed datasets have their data spread across multiple servers, but each server contains not only its chunk of the data, but also backup copies of every other chunk. In theory, all but one of the servers could fail, and users could still query the entire dataset (although much more slowly than if all the servers were available). Such a dataset is highly fault tolerant.


## Question 5

What is MapReduce? How is it related to Hadoop?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________


MapReduce is a technique used in Hadoop and Spark to query large, distributed datasets. When a user constructs a query, it is first 'mapped' to every server in the distributed dataset (i.e., each server is asked to perform the same query at the same time). Once each server has run its query, the results are then aggregated ('reduced') to yield results that represent the entire dataset as a whole.
