# Extra practice with library data

Here I'm doing some extra practice with data from the IMLS (Institute of Museum and Library Services). You can find the data and documentation here: https://www.imls.gov/research-evaluation/data-collection/public-libraries-survey.

```python
# Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Read in the data
df = pd.read_excel('PLS_FY17_AE_pud17i.xlsx')
df.head()
```

```python
df.info()
```

```python
# col_list = list(df.columns)
for col in df.columns:
    print(col, type(df[col][3]))
```

```python
by_state = df.groupby('STABR')
```

```python
state_mean_salaries = pd.DataFrame(by_state['SALARIES'].mean())
state_mean_salaries.reset_index(inplace=True)
```

```python
state_mean_salaries = state_mean_salaries.sort_values('SALARIES', ascending=False)
state_mean_salaries
```

```python
plt.figure(figsize=(16,10))
sns.barplot(y=state_mean_salaries.STABR, x=state_mean_salaries.SALARIES, 
            orient='h', color='cadetblue')
plt.title('Mean expenditure on salaries per library by state (RAW DATA)')
plt.ylabel('State')
plt.xlabel('Mean expenditure on salaries (10^7)')
plt.show();
```

```python
state_mean_collex = pd.DataFrame(by_state['PRMATEXP'].mean())
state_mean_collex.reset_index(inplace=True)
state_mean_collex.sort_values('PRMATEXP', ascending=False, inplace=True)
state_mean_collex.head()
```

```python
plt.figure(figsize=(16,10))
sns.barplot(y=state_mean_collex.STABR, x=state_mean_collex.PRMATEXP, 
            orient='h', color='cadetblue')
plt.title('Mean expenditure on collections per library by state (RAW DATA)')
plt.ylabel('State')
plt.xlabel('Mean expenditure on collections')
plt.show();
```

```python
state_mean_digcol = pd.DataFrame(by_state['ELMATEXP'].mean())
state_mean_digcol.reset_index(inplace=True)
state_mean_digcol.sort_values('ELMATEXP', ascending=False, inplace=True)
state_mean_digcol.head()
```

```python
plt.figure(figsize=(16,10))
sns.barplot(y=state_mean_digcol.STABR, x=state_mean_digcol.ELMATEXP, 
            orient='h', color='cadetblue')
plt.title('Mean expenditure on digital collections per library by state (RAW DATA)')
plt.ylabel('State')
plt.xlabel('Mean expenditure on digital collections')
plt.show();
```

```python
by_state_bkmobile = pd.DataFrame(by_state.BKMOB.sum())
by_state_bkmobile.reset_index(inplace=True)
by_state_bkmobile.sort_values('BKMOB', ascending=False, inplace=True)
by_state_bkmobile.head()
```

```python
plt.figure(figsize=(16,10))
sns.barplot(y=by_state_bkmobile.STABR, x=by_state_bkmobile.BKMOB, 
            orient='h', color='orange')
plt.title('Number of bookmobiles per state (RAW DATA)')
plt.ylabel('State')
plt.xlabel('Bookmobiles')
plt.show();
```

```python
bkmobile_clean = df[df.BKMOB > 0]
len(bkmobile_clean)
```

```python
bkmobile_clean_by_state = pd.DataFrame(bkmobile_clean.groupby('STABR')['BKMOB'].sum())

bkmobile_clean_by_state.reset_index(inplace=True)
bkmobile_clean_by_state.sort_values('BKMOB', ascending=False, inplace=True)
bkmobile_clean_by_state.head()
```

```python
sns.set_context('talk')
```

```python
plt.figure(figsize=(16,16))
sns.barplot(y=bkmobile_clean_by_state.STABR, x=bkmobile_clean_by_state.BKMOB, 
            orient='h', color='orange')
# plt.suptitle('')
plt.title('Number of bookmobiles per state\n(Includes only states reporting >0 bookmobiles)')
plt.ylabel('State')
plt.xlabel('Bookmobiles')
plt.show();
```
